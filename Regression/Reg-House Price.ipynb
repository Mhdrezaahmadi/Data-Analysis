```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\Machine learning\Data\house_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
x = np.asanyarray(df[['sqft_living']])
y = np.asanyarray(df[['price']])
```

# Modeling

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state= 4)
print ('Trainset Size', x_train.shape, y_train.shape)
print('Testset Size', x_test.shape, y_test.shape)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
print('Coefficient:', regr.coef_)
print('Intercept', regr.intercept_)
```

```python
plt.scatter(df.sqft_living, df.price , color='black')
plt.plot(x_train, regr.coef_[0][0]* x_train + regr.intercept_[0], '-b')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price of the House')
```

```python
yhat = regr.predict(x_test)
yhat
```

# Evaluation

```python
from sklearn import metrics 
print('MAE', metrics.mean_absolute_error(y_test, yhat))
print ('MSE', metrics.mean_squared_error(y_test, yhat))
```
